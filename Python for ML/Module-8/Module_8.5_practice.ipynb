```python
class Vehicle:
    def __init__(self, vehicleType, name, speed):
        self.__vehicleType = vehicleType
        self.__name = name
        self.__speed = speed

    def type_getter(self):
        return self.__vehicleType
    def name_getter(self):
        return self.__name
    def speed_getter(self):
        return self.__speed
        
    def classify(self):
        if self.__speed > 100:
            return "High speed vehicle"
        else:
            return "Normal Vehicle"

    def show_info(self): 
        print(f"Type: {self.type_getter()}, Name: {self.name_getter()}, Speed: {self.speed_getter()} km/h, Predicted: {self.classify()}")


class Car(Vehicle):
    def __init__(self, vehicleType, name, speed):
        super().__init__(vehicleType, name, speed)

    def show_info(self): 
        print(f"Type: {self.type_getter()}, Name: {self.name_getter()}, Speed: {self.speed_getter()} km/h, Predicted: {self.classify()}")

class Bike(Vehicle):
    def __init__(self, vehicleType, name, speed):
        super().__init__(vehicleType, name, speed)

    def show_info(self): 
        print(f"Type: {self.type_getter()}, Name: {self.name_getter()}, Speed: {self.speed_getter()} km/h, Predicted: {self.classify()}")




n = int(input("Number of vehicles"))

lst = []

for i in range(n):
    x = input("What type of vehicle: ")
    y = input("Vehicle name: ")
    z = int(input("Vehicle's max speed"))
    if x.lower() == "car":
        cr = Car(x, y, z)
        lst.append(cr)
    else:
        bk = Bike(x, y, z)
        lst.append(bk)

print()
for i in range(n):
    print(f"{lst[i].type_getter()} Added: {lst[i].name_getter()}")


print()
print("--- Vehicle Details ---")
print()

for a in lst:
    a.show_info()
        
```
