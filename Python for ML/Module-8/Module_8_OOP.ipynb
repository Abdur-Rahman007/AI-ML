```python
# Create class
# Base class / Parent class

class Phone:
    category = "Electronics" # Class Attribute

    # Constructor
    def __init__(self, model, battery, camera, battery_percentage = 100):
        self.model = model                #Instance Attribute
        self.battery = battery
        self.camera = camera
        self.battery_percentage = battery_percentage

    # Methods
    def charge_left(self):
        print(f"The battery percentage {self.battery_percentage}%")

    def capture(self, photo):
        if self.battery_percentage <= 0:
            print("No charge")
        else:
            self.battery_percentage -= photo
            print(f"Photo captured in {self.model} {photo} photos")
    def charge_phone(self, hour):
        print(f"The phone is charging for {hour} hours")
        if(self.battery_percentage <100):
            self.battery_percentage = min(100, self.battery_percentage+hour)

class Cooling_Mechanism:
    def __init__(self, cooling_method):
        self.cooling_method = cooling_method

    def cooling_on(self):
        print(f"The ssystem is being cooled by {self.cooling_method}")

        
```

```python
# create object

apple = Phone("Iphone 17", 5000, 50)
apple.category = "Super Electronics"  # Overwrite

samsung = Phone("Samsung S25", 5500, 40)
nokia = Phone("A50", 6000, 30)

print(apple.category)
print(apple.camera)

print(samsung.category)
print(samsung.battery)

print(nokia.category)
print(nokia.model)

apple.capture(10)
apple.charge_left()
```

```python
#Inheritance
# Child class / derived class

class SmartPhone(Phone):
    def __init__(self, model, battery, camera, processor):
        super().__init__(model, battery, camera)
        self.processor = processor
    def charge_phone(self, hour):
        print("Super charging mode is on.")
        super().charge_phone(hour)
        super().charge_left()
```

```python
pro = SmartPhone("X", 5000, 30, "AMD")

print(pro.model) # instance inheritade
pro.charge_left()  # Method inheritade
pro.capture(20)
pro.charge_left()
pro.charge_phone(25)
```

```python
# Multiple inheritance

class SmartPhone_CoolingMode(SmartPhone, Cooling_Mechanism):
    def __init__(self, model, battery, camera, processor, cooling_method):
        SmartPhone.__init__(self, model, battery, camera, processor)
        Cooling_Mechanism.__init__(self, cooling_method)

```

```python
pro_cooling = SmartPhone_CoolingMode("Y", 5000, 100, "SD", "Nitrogen")

print(pro_cooling.processor)  # from SmatPhone Class
print(pro_cooling.battery)   # from Phone Class which was inherited by SmartPhone Class
print(pro_cooling.cooling_method) # from Cooling_Mechanism Class
pro_cooling.cooling_on() # Method of Cooling_Mechanism
pro_cooling.charge_left() # Method of Phone Class which was inherited by SmartPhone Class
```

```python
# Polymorphsim

class Camera:
    def __init__(self, name):
        self.name = name

    def capture(self):
        print("a photo is captured")

class Smart(Camera):
    def __init__(self, name, resolution):
        super().__init__(name)
        self.resolution = resolution

    # Method Overrideing (same method is in the parent class)
    def capture(self):
        print("Photo is captured by Phone")

class DSLR(Camera):
    def __init__(self, name, resolution):
        super().__init__(name)
        self.resolution = resolution
        
    # Method Overrideing (same method is in the parent class)
    def capture(self):
        print("Photo is captured by DSLR")

class Drone(Camera):
    def __init__(self, name, resolution):
        super().__init__(name)
        self.resolution = resolution

    # Method Overrideing (same method is in the parent class)
    def capture(self):
        print("Photo is captured by Drone")
        

phone = Smart("Phone", 30)
dslr = DSLR("DSLR", 200)
drone = Drone("Drone", 150)

phone.capture()
dslr.capture()
drone.capture()

```

```python
# Encapsulation
class Mobile:
    def __init__(self, name, model, imei, owner):
        self.name = name
        self.__model = model # Private
        self.__imei = imei # Private
        self.__owner = owner ## Private

        
    def charge(self):
        print("Phone is charging.")

    def imei_getter(self):
        return self.__imei

    def model_getter(self):
        return self.__model

    def owner_getter(self):
        return self.__owner

    def owner_setter(self, owner_name):
        self.__owner = owner_name
        print("Owner successfully changed")


iphone = Mobile("Phone", "17", "1xM4c", "AR")

print(iphone.name)

iphone.name = "PH"   #changeable cz it is public
print(iphone.name)

#print(iphone.imei) # can't access directly cz private
print(iphone.imei_getter())

iphone.imei = "123"   # not changeable directly
print(iphone.imei_getter())

## accessing and changing Pivate attribute
print(iphone.owner_getter())
iphone.owner_setter("Ratul")
print(iphone.owner_getter())
```

```python
# Abstraction

from abc import ABC, abstractmethod

class Telephone(ABC):  #ABC -> Abstract Base Class (here telephone is a abstract class)
    
    @abstractmethod # declaring a abstract method (and it have to be empty)
    def make_call(self):
        pass
        

class Sphone(Telephone):  # If a abstract class is inherited the chld must have to implement the abstract methods
    def make_call(self):
        print("Making a call using Sphone.")
    
class Iphone(Telephone):
    def make_call(self):
        print("Making a call using Iphone.")


ip = Iphone()

ip.make_call()
```
