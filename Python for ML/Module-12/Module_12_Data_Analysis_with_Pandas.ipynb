```python
# Changing column name, inplace=True means change in the main table
df.rename(columns = {'FullName':'Full Name', 'Algorithm Marks':'Algo Marks'}, inplace=True)
df
```

```python
# Dropping rows and columns,Assigning values and Iteration

df.drop(0, inplace=True) # droping the index 0 row

df
```

```python
df.drop('Instructor', axis=1, inplace=True)  # Droping the Instructor column
df
```

```python
 # Assign value
df.loc[1, 'Python Marks'] = 90
df.loc[1, 'CompletionStatus'] = 'Completed'

df.head(5)
```

```python
df.loc[1:3, 'Python Marks'] += 2
df.head(5)
```

```python
  # Iteration (iteration is not prefarable, broad casting is preferable)

for i, series in df.iterrows():  # i-> index, series-> rows
    print(f"{i}: {series}")
```

```python
  # iteration
for i in df.itertuples(index=False):  # false is written to print without index
    print(i)
```

```python
# Sorting (Ascending)

copy = df.sort_values('Data Structure Marks')  # creating a copy not changing in main, to change in main use inplace
copy   # sorted accoring "Data Structure Marks"
```

```python
# Sorting (Descending)
copy2 = df.sort_values('Data Structure Marks', ascending=False)
copy2
```

```python
# Sorting according multple values
# Sort (Descending)
copy3 = df.sort_values(['Data Structure Marks', 'Python Marks'], ascending=False)
copy3
```

```python
# Sorting accordng multiple value and set diffrent orders 
# Sort (Descending and ascending)
copy4 = df.sort_values(['Data Structure Marks', 'Python Marks'], ascending=[0,1]) # 0->descending, 1->ascending 
copy4
```
