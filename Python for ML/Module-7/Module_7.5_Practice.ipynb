```python
# Problem - 1:
a = set(input().split())
b = set(input().split())

total_friends = len(a.union(b))

mutual_friends = a.intersection(b)

unique_in_a = a.difference(b)
unique_in_b = b.difference(a)

print(f"""Mutual friends: {mutual_friends}
Unique to A: {unique_in_a}
Unique to B: {unique_in_b}
Total unique friends: {total_friends}
""")
```

```python
# Problem - 2:
data = input().split()

students = []
for i in range(0, len(data), 2):
    name = data[i]
    marks = int(data[i+1])
    students.append((name, marks))

students.sort(key = lambda x: x[1], reverse=True)

print("Top 3 Students:")
for i in range(min(3,len(students))):
    print(f"{students[i][0]} - {students[i][1]}")
```

```python
# Problem - 3

words = input().split()

ans = list(filter(lambda x: len(x)>4, words))

print(ans)
```

```python
# problem-4

num = list(map(int, input().split()))

sqr = list(map(lambda x: x**2, filter(lambda x: x%2==0, num)))
print(sqr)
```
