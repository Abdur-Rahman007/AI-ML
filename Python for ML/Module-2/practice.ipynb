```python
# CF problem https://codeforces.com/group/MWSDmqGsZm/contest/219432/problem/C

n = int(input())

inp = input()

num = inp.split()

even = 0
odd = 0
pos = 0
neg = 0

for i in range(n):

    x = int(num[i])
    
    if x>0:
        pos += 1
    elif x<0:
        neg += 1
    if x%2 == 0:
        even += 1
    elif x%2 == 1:
        odd += 1
        
print("Even:", even)
print("Odd:", odd)
print("Positive:", pos)
print("Negative:", neg)

```
