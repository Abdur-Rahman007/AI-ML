```python
#Dictionary in python (Key, alue)

dic = {"Name": "Ratul", "Age": 20, "Address": "Dhaka", "Number": [10, 20, 30]}
print(type(dic), dic)

print(dic["Name"])
print(dic.get("Age"))

#changing value
dic["Name"] = "Abdur Rahman"
print(dic["Name"])

#If value dose not exist
print(dic.get("Roll"))   #Print's None
print(dic.get("Roll", 0)) #Print's 0
```

```python
# Store new values in dictionary
dic2 = {}
dic2["Class"] = 7
dic2["Room"] = 5002
dic2["Sec"] = "C"

print(dic2)

del dic2["Room"]
print(dic2)
```

```python
# dictionary into dictionary

dic3 = {"key" : {"name": "AR", "Age":23}}
print(dic3)

print(dic3["key"]["name"])
```

```python
# Dictionary View Method
dic4 = {"Name": "Ratul", "Age": 20, "Address": "Dhaka", "Number": [10, 20, 30]}

keys = dic4.keys()
print(keys)
values = dic4.values()
print(values)

## items() method returns all the key:values in a list and all key:values are kept in a tuple
items = dic4.items()
print(items)
```

```python
# Iteration 

for key,value in dic4.items():
    print(key, value)
```

```python
# Dictionary Comprehension (expression iterable confition[if any])
squre = {x:x**2 for x in range(10) if x%2==0}

print(squre)

#Mergeing 2 data as key value in a dictionary
co_ordinates = [(10, 10.5), (25.5, 25), (8, 8.5)]
locations = ["Dhaka", "Sylhet", "Cumilla"]
exact_location = {x:y for x,y in zip(co_ordinates, locations)}
print(exact_location)
```
