```python
x = 100
y = 3.14
z = "Abdur rahman"
t = True

print("Type           Value")
print(f"{type(x)} -> {x}")
print(f"{type(y)} -> {y}")
print(f"{type(z)} -> {z}")
print(f"{type(t)} -> {t}")
```

```python
## input

age = input("What is your age")

print(age, type(age))

age = int(age)   ## Type casting
print(age, type(age)) 
```

```python
# Direct type casting

num = float(input("Give a floating number"))

print(num, type(num))

num = num + 0.2

print(num)
```

```python
#Operators

x = 10 ; y = 3

sum = x + y
sub = x - y
mult = x*y
div = round(x/y, 2)
int_div = x//y
rem = x%y

print(sum, sub, mult, div, int_div, rem)
```

```python
#Logical Operator

x = 10; y = 7; z = 15

result = (x>y) and (y>z)
print(result)

ans = (x>y) or (y>z)
print(ans)
```
