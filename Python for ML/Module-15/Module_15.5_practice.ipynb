```python
import pandas as pd

data = {
    "EmployeeID": ["E101","E102","E103","E101","E104","E105","E106","E107","E108","E109","E102"],
    "EmployeeName": ["Alice","Bob","Charlie","Alice","David","Eve","Frank","Grace","Heidi","Ivan","Bob"],
    "Age": [25,30,22,25,45,29,None,34,38,26,30],
    "Salary": [50000,None,52000,50000,80000,58000,75000,None,None,53000,None],
    "Price": [1200,1500,1100,1200,2500,1300,2200,1900,1800,1400,1500],
    "Marks1": [85,78,92,85,88,76,90,81,84,95,78],
    "Marks2": [90,82,95,90,85,79,91,83,86,92,82],
    "Marks3": [88,80,90,88,89,81,92,80,88,94,80]
}

df = pd.DataFrame(data)
print(df)
```

```python
df['Discounted_Price'] = df['Price'] - (df['Price']*0.1)
df
```

```python
df.to_csv('discounted_products.csv')
```

```python
df = pd.read_csv('discounted_products.csv')
df
```

```python

ans = df['EmployeeID'].duplicated().sum()
print(ans)
```

```python
df.drop_duplicates(subset=['EmployeeID'], inplace=True)
```

```python
df
```

```python
age_null = df['Age'].isnull().sum()
print("Number of Age null values is",age_null)

salary_null = df['Salary'].isnull().sum()
print("Number of Salaty null values is", salary_null)
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df
```

```python
df.dropna(subset=['Salary'], inplace=True)
df
```

```python
M_1_mean = df['Marks1'].mean()
M_1_max = df['Marks1'].max()

M_2_mean = df['Marks2'].mean()
M_2_max = df['Marks2'].max()

M_3_mean = df['Marks3'].mean()
M_3_max = df['Marks3'].max()

print("Mean of Marks1", M_1_mean)
print("Max of Marks1", M_1_max)

print("Mean of Marks2", M_2_mean)
print("Max of Marks2", M_2_max)

print("Mean of Marks3", M_3_mean)
print("Max of Marks3", M_3_max)
```

```python
df['Total Marks'] = df['Marks1'] + df['Marks2'] + df['Marks3']
df
```

```python
print(df['Total Marks'].std())
```
