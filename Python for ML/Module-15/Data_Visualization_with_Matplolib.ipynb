```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 2D Line Plot
hours = [2, 3, 4, 1, 5, 7, 3]
days = [1, 2, 3, 4, 5, 6, 7]

plt.plot(days, hours)
```

```python
# Add label of x and y axis and add a title

plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.title("Trend of a student study time over a week")

plt.plot(days, hours)
```

```python
# 2D line from pandas DF

df = pd.read_csv('enrollment_data.csv')
df
```

```python
plt.xlabel('Years')
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 10 years')

plt.plot(df['Year'], df['Programming'])
```

```python
# Ploting for multiple data
plt.xlabel('Years')
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 10 years')

plt.plot(df['Year'], df['Programming'])
plt.plot(df['Year'], df['Digital Marketing'])
```

```python
# Adding label of the curve line
plt.xlabel('Years')
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 10 years')

plt.plot(df['Year'], df['Programming'], label = 'Programming')
plt.plot(df['Year'], df['Digital Marketing'], label = 'Digital Marketing')

plt.legend()  # default location 'best'.
```

```python
# Customizing Line plot ( color,linestyle,marker) and Grid

  # Adding Some Visualization
  
plt.xlabel('Years')
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 10 years')

plt.plot(df['Year'], df['Programming'], label = 'Programming', color='#17d106', linewidth=4, linestyle='dotted', marker='o') # Hex color
plt.plot(df['Year'], df['Digital Marketing'], label = 'Digital Marketing', color='black', linestyle='dashed', marker='o', markersize = 10)   # maeker points the plot

plt.legend() 
```

```python
# Grid(clear Visualization)
  
plt.xlabel('Years')
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 10 years')

plt.plot(df['Year'], df['Programming'], label = 'Programming', color='#17d106', linewidth=4, linestyle='dotted', marker='o') # Hex color
plt.plot(df['Year'], df['Digital Marketing'], label = 'Digital Marketing', color='black', linestyle='dashed', marker='o', markersize = 10)   # maeker points the plot

plt.legend() 
plt.grid()
plt.show()
```

```python
# Bar Chart (Chetegorizing)

data = {
'Student': ['A', 'B', 'C', 'D', 'E', 'F', 'G'],
'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
'StudyHours': [6, 2, 5, 3, 7, 4, 8, ],
'Attendance': [92, 60, 88, 70, 95, 80, 98],
'Grade': ['Pass', 'Fail', 'Pass', 'Fail', 'Pas s', 'Pass', 'Pass']
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# gender wise chetegorize

gender_group = df.groupby('Gender').size()

print(gender_group.index)
gender_group.values
```

```python
plt.bar(gender_group.index, gender_group.values, color=['red', 'green'], edgecolor='black')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Bar Chart of male and female count')
```

```python
# Bar chart on real data
dframe = pd.read_csv(r'..\student_data.csv')
dframe
```

```python
status = dframe.groupby('CompletionStatus').size()

plt.bar(status.index, status.values, edgecolor='black', color=['Green', 'orange', 'red'])

plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Bar Chart on Completion Status')
```

```python
status = dframe.groupby('Location').size()

plt.bar(status.index, status.values, edgecolor='black')

plt.xlabel('Location')
plt.ylabel('Count')
plt.title('Bar Chart on Location')
```

```python
# Pie Chart
status = dframe.groupby('CompletionStatus').size()
plt.pie(status)
plt.show()
```

```python
# Customize Pic chart
status = dframe.groupby('CompletionStatus').size()
plt.pie(status, labels=status.index, autopct='%1.1f%%', explode=(0.1,0,0), shadow=True) # autopct = auto percentage
plt.show()
```
