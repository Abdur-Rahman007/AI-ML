```python
##Function(Empty parameterized function)
def greet():
    print("Hello")
greet()

##parameterized
def fun(name):
    print(f"Hello, {name}")
fun("Ratul")                # Must have to give argument in the calling time

## Parameterized function default value is set(it can be call by argument or empty)
def fun1(game = "Cricket"):
    print(f"His fav game is {game}")
fun1()
fun1("Football")

```

```python
## Positional Argument (the parameters and arguments positions are connected according to the position)

def ope(a, b, c):
    a = a**2
    b = b**2
    return a+b+c
ans = ope(10, 20, 30) + 10

print(ans)


##Keyword Argument (the parameters and arguments positions are connected through declaration)

def ope1(a, b, c):
    a = a**2
    b = b**2
    return a+b+c
ans1 = ope1(b=10, c=20, a=30) + 10

print(ans1)

## Variable sized argument (Parameters are not limited and argumrnts are passed as tuple )
def Variable_Sized(*args):
    s = 0
    for i in args:
        s += i**2
    return s

ans2 = Variable_Sized(10, 20, 30, 40, 50)

print(ans2)

## Variable_Sized key_word argument (Parameters are not limited and argumrnts are passed as dictionary{key value} )
def student(**kwargs):
    for key,val in kwargs.items():
        print(f"{key} {val}")

student(name="Ratul", cls=10, roll=2, marks=90.5)
```

```python
#return

def one():
    return "Ratul"

print(type(one()))

def two():
    return 100

print(type(two()))

def three():
    return 50.5

print(type(one()))

def four():
    return True

print(type(four()))


## return unpacking

def upaking():
    return 10, 50

x, y = upaking()

print(f"{x} {y}")


def upaking2():
    a = [10, 20, 30]
    b = [40, 50, 60]

    return a, b

c, d = upaking2()
print(c)
print(d)
```

```python
##Genarator (give a chunk of data as the size you want. Not the whole data at a time)

lst = [x for x in range(500)] #dataset

def data_loader(chunk_size, lst):
    for i in range(0,len(lst),chunk_size):
        yield lst[i:i+chunk_size]

x = data_loader(5, lst)
print(type(x))

print(next(x))
print(next(x))
    
```

```python
## Lambda Function (is a way to write a method in one line)

def square_addition(x, y):  #Normal way
    return x**2 + y**2

print(square_addition(2,3))

#1
sq_add = lambda x,y : x**2 + y**2   # in Lambda ( lambda parameter : expression )

print(sq_add(2, 3))

#2
if_even = lambda a : a%2 == 0

print(if_even(10))
```

```python
## Map Function (using map function we can use a function in every element of a list)

#List
lst = [1, 2, 3, 4, 5]

def square(a):
    return a**2

lst = list(map(square, lst))  ## map(function, iterable)

print(lst)


# String
s = "This is the time of AI"

st = list(map(str.upper, s))

print(st)

st = "".join(st)
print(st)
```

```python
#Map with Lambda function

lst1 = [1, 2, 3, 4, 5, 6]

lst2 = list(map(lambda x : x**2, lst1))

print(lst2)
```

```python
# Filter Function (filter(function, iterable)-> function have to return True or False)

number = [x for x in range(100)]

#1
even = list(filter(lambda x : x%2==0, number))
print(even)

#2
fifty = list(filter(lambda x: x>50, number))
print(fifty)

#3(We can also filter None like this)
data = [1, 2, "Hello", [], [1,2], None, 30.5, None, True, False]

clean_data = list(filter(None, data))
print(clean_data)
```

```python
# Filter in string
#1
s1 = "This is the time of AI"

st1 = list(map(str.upper, s1))

print(st1)

vowel = list(filter(lambda x : x in "AEIOU", st1))
print(vowel)

#2
fruit = ["aa", "bbbb", "cc", "dddd"]

filtering = list(filter(lambda x: len(x)>2, fruit))
print(filtering)
```

```python
# Reduce Function (reduce function returns only one answer)

from functools import reduce

num = [1, 2, 3, 4, 5, 6]

#1
summation = reduce(lambda x,y: x+y, num)

print(summation)

#2
maximum = reduce(lambda x,y: x if x>y else y, num)  ## <value_if_true> if <condition> else <value_if_false>

print(maximum)

#3
lst9 = ["Hello", "Every", "one", "It's", "Ratul"]

string = reduce(lambda x,y: x+" "+y, lst9)
print(string)
```

```python
# Min and max value of a list

from functools import reduce

n = int(input())
lst = list(map(int, input().split()))

mn = reduce(lambda x, y:x if x<y else y, lst)
mx = reduce(lambda x,y : x if x>y else y, lst)
print(f"{mn} {mx}")
```

```python
#Print the average(mean) of the array with 6 digits after the decimal point.

from functools import reduce

n = int(input())
lst = list(map(float, input().split()))

summation = reduce(lambda x,y : x+y, lst) / n
print(f"{summation:.7f}")
```
