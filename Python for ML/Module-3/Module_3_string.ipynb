```python
#Indexing and Slicing
str = """Lorem is simply dummy text of the printing and typesetting industry. 
Lorem Ipsum has been the industry's standard dummy text ever since the 1500s."""

print(str[0])

slic = str[0:23] #[start:end:steps]
print(slic)
print()
print("1.",slic[-1])
print("2.",slic[-2])
print("3.",slic[-3])
print("4.",slic[-4])
print("5.",slic[-5])
```

```python
## String Method

string = "I have an Alexa, Alexa"

str_process = string.lower()

print(str_process)
print(len(string))

print("Alexa" in str_process)

print(str_process.find("alexa"))
print(str_process.rfind("alexa"))

#count the sub string number

print(str_process.count("alexa"))

# Replace a sub string
str2 = string.replace("Alexa","AI")
print(str2)
```

```python
## spliting, joining, formatted string

prompt = "What is phitron ?"
token = prompt.split()
print(token)

sentence = "_".join(token)
sentence2 = " ".join(token)
print(sentence)
print(sentence2)
```
