```python
## List

numbers = [100, 20, 40, 80, 70]

print(numbers[2])

numbers[2] = 30
print(numbers[2])
print(numbers)

# slicing
new_list = numbers[0:3]
print(new_list)
```

```python
## Adding item
numbers.append(50)
print(numbers)

## Add item at any position
numbers.insert(0, 45)
print(numbers)
```

```python
## pop
numbers.pop()
print(numbers)

#remove
numbers.remove(80)
print(numbers)
```

```python
## Sorting
numbers.sort()
print(numbers)

numbers.sort(reverse = True) # or numbers.reverse()
print(numbers)
```

```python
# List as Stack
stack = []

##insert in stack
stack.append(1)
stack.append(2)
stack.append(3)
stack.append(4)
stack.append(5)

## printng and deleting the top element from stack
print(stack.pop())
print(stack.pop())
print(stack.pop())
print(stack.pop())
print(stack.pop())
```

```python
# List as Queue
queue = []

##insert in queue
queue.append(1)
queue.append(2)
queue.append(3)
queue.append(4)
queue.append(5)

## printng and deleting the top element from 
print(queue)
print(queue.pop(0))
print(queue)
print(queue.pop(0))
print(queue)
print(queue.pop(0))
print(queue)
print(queue.pop(0))
print(queue)
print(queue.pop(0))
print(queue)
```

```python
#List comprihension
even = []
for i in range(1,61):
    if i%2 == 0:
        even.append(i)
print(even)

#1
odd = [x for x in range(1, 61) if x%2 != 0] # [expression lop condition(if any)]
print(odd)

#2
fruits = ["Apple", "Orange", "ChiLLi"]
all_upper = [fruit.upper() for fruit in fruits]
print(all_upper)
```
