```python
import numpy as np
import pandas as pd

data = {
    'Room_Size' : [500,1000,1500,2000,2500],
    'Bath_Size' : [40,60,80,100,120]
}

df = pd.DataFrame(data)

df
```

```python
covariance_matrix = np.cov(df,rowvar=False)

# print(type(covariance_matrix))


eigen_values,eigen_vector = np.linalg.eig(covariance_matrix)

# print(eigen_values)
# print(eigen_vector)


selected_vector = eigen_vector[::,0:1]

print(selected_vector)



df['house_size'] = np.dot(df,selected_vector)

df
```
