```python
# 2-> SD or varience which one is easy to interupt
temp =[29, 31, 33, 33, 32, 31, 30]
tmp = pd.DataFrame({'Temperature': temp})
```

```python
deviation = np.std(tmp['Temperature'])
variance = np.var(tmp['Temperature'])
print("Standard Deviation", deviation)
print("Variance", variance)

# Standard deviation is easy to interupt cz SD have the same unit as the data celcius(C). But Variance have C^2.
```

```python
cat_weight = [ 2.5, 3.0, 3.2, 3.3, 3.4, 3.5, 3.6, 3.9, 4.0, 4.5, 15]

Q1 = np.percentile(cat_weight, 25)
Q2 = np.percentile(cat_weight, 50)
Q3 = np.percentile(cat_weight, 75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in cat_weight if (x>upper_bound or x<lower_bound)]
print("Percentile P25: ", Q1)
print("Percentile P50: ", Q2)
print("Percentile P75: ", Q3)
print("IQR: ", IQR)
print("Outlier Range:", lower_bound, "to", upper_bound)
print("outliers", outliers)
```

```python
# 3- > Conditional Probabilaty
"""1. In a survey of 100 people:
40 like pizza (A)
50 like burgers (B)
20 like both (A ∩ B)
 Find P(A), P(B), P(A ∩ B), and P(A | B).
"""

PA = 40/100
PB = 50/100
PAB = 20/100

print("P(A)", PA)
print("P(B)", PB)
print("P(AB)", PAB)

PAgivenB = PAB/PB
print("P(A|B)", PAgivenB)
```

```python
# Bayes Therom
"""
2. Email dataset of 1 000 mails:
100 spam (S)
40 of these contain “free” (F)
60 non-spam also contain “free”
 Compute P(S | F) using Baye's Theorem →
 
 Explain what this means for a spam filter.

"""

PS = 100/1000
PF = 100/1000

PFgivenS = 40/100

bayes_PSgivenF = (PFgivenS*PS)/PF

print(bayes_PSgivenF)
```

```python
PD = 1/10000
PmD = 1-PD          # P(-D)
PpbD = 0.99         # P(+|D)
PmbmD = 0.99       # P(-|-D)
PpbmD = 1 - PpbD  # P(+|D) + P(+|-D) = 1
Pp = ((PpbD*PD)+(PpbmD*PmD))  # P(+) = P(+|D)*P(D) + P(+|-D) P(-D)

PDbp = (PpbD*PD) / Pp

print("P(D|+)=",PDbp)
```

```python
# Part - c

TP = 120
FP = 80
FN = 30
TN = 770
total = 1000

accuracy = (TP+TN)/total
precision = TP/(TP+FP)
recall = TP/(TP+FN)
F1 = (2*precision*recall)/(precision+recall)
specifity = TN/(TN+FP)
NPV = TN/(TN+FN)
prevalence = (TP+FN) / total


print("Accuracy=", accuracy)
print("Precision=", precision)
print("Recall=", recall)
print("F1=", F1)
print("Specifity=", specifity)
print("Negative Predictive Value=", NPV)
print("Prevalence", prevalence)
```
