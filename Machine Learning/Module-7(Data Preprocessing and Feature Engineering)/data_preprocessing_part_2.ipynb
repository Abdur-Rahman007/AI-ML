```python
# Age have missing values. So to work for it at first we have to understand its nature
plt.figure(figsize=(8,6))
sns.histplot(df_titanic['Age'], kde = True, bins=40) # kde draws the wave signal
plt.title("Age Distribution of Titanic Passenger")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
# We will use Median as the feautre value are right skewed
```

```python
print("Distribution of Embarked Column: ")
plt.figure(figsize=(8,6))
sns.countplot(data=df_titanic, x='Embarked')
plt.title("Embarked distributon of Titanic Passenger")
plt.xlabel("Embarked")
plt.ylabel("Passenger Frequency")
plt.show()
```

```python
# 1. Handle numerical missing values: Age
age_median = df_titanic['Age'].median()
df_titanic['Age'] = df_titanic['Age'].fillna(age_median)

# 2. Handle categorical missing values: Embarked
embarked_mode = df_titanic['Embarked'].mode()[0]
df_titanic['Embarked'] = df_titanic['Embarked'].fillna(embarked_mode)

# 3. Drop Cabin (too many missing values)
df_titanic = df_titanic.drop(columns=['Cabin'])
```

```python
#  Encoding Categorical Variables or Features
df_heart = pd.read_csv("heart.csv")

display(df_heart.head(10))

display(df_heart.dtypes) #object type means categorical data
```

```python
# checking how many categories are there in a column

categorical_cols = ["Sex", "ChestPainType", "RestingECG", "ExerciseAngina", "ST_Slope"]

for c in categorical_cols:
    plt.figure(figsize=(5,4))
    df_heart[c].value_counts().plot(kind='bar')
    plt.title(f"Value counts for {c}")
    plt.ylabel("Count")
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Label Encoding for binary categorical columns -> Sex and ExerciseAngina
le = LabelEncoder()
df_heart["Sex"] = le.fit_transform(df_heart["Sex"])
df_heart["ExerciseAngina"] = le.fit_transform(df_heart["ExerciseAngina"])
```

```python
df_heart.head(10)
```

```python
# OneHot Encoding for nominal categorical columns (Not binary)
cat_cols = ["ChestPainType", "RestingECG", "ST_Slope"]

df_heart_encoded = pd.get_dummies(
    df_heart,
    columns= cat_cols,
    dtype= int
)
```

```python
df_heart_encoded.head(10) # All datas are encoded to numerical values no catagorical 
```

```python
# Normalization and scaling
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler

tar_col = "HeartDisease"

# spliting target col and feature value
x = df_heart_encoded.drop(columns=[tar_col])
y = df_heart_encoded[tar_col]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42) # what is test_set and what is train set. how test_size & random_state code is working here

# Standard Scaling
scaler_sd = StandardScaler()
x_train_std = scaler_sd.fit_transform(x_train) # mean snd SD will be calculated from x_train
x_test_std = scaler_sd.transform(x_test)

# MinMax Scaling
scaler_mm = MinMaxScaler()
x_train_mm = scaler_mm.fit_transform(x_train) # mean snd SD will be calculated from x_train
x_test_mm = scaler_mm.transform(x_test)

print("-----Display Standard Scaling Data-----")
# Convert scaled arrays back to DataFrame for better visualization with column names
x_train_std_df = pd.DataFrame(x_train_std, columns= x_train.columns, index = x_train.index)
x_test_std_df = pd.DataFrame(x_test_std, columns= x_test.columns, index = x_test.index)

display(x_train_std_df.head())
display(x_test_std_df.head())

print("-----Display Min-Max Scaling Data-----")
# Convert scaled arrays back to DataFrame for better visualization with column names
x_train_mm_df = pd.DataFrame(x_train_mm, columns= x_train.columns, index = x_train.index)
x_test_mm_df = pd.DataFrame(x_test_mm, columns= x_test.columns, index = x_test.index)

display(x_train_mm_df.head())
display(x_test_mm_df.head())
```
