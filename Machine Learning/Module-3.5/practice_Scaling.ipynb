```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
b1 = pd.DataFrame({'H' : [150, 160, 170, 175, 180],
    'W' : [58, 62, 65, 66, 190]
})
b1
```

```python
# Min-Max scaling
mn = b1.min()
mx = b1.max()
rg = mx - mn

mm = (b1-mn)/rg
mm.round(2)
```

```python
# Standardize (Z-Scaling)
mean = b1.mean()
sd = b1.std()
z = (b1-mean)/sd
z.round(2)
```

```python
# Robust scaling
median = b1['W'].median()
q1 = b1['W'].quantile(0.25)
q3 = b1['W'].quantile(0.75)
iqr = q3 - q1

rb = (b1['W']-median)/iqr
rb.round(2)
```

```python
# One Hot encoding

b2 = pd.DataFrame({'Cities' : ['Dhaka', 'Chattogram', 'Dhaka', 'Rajshahi', 'Rajshahi']
})

c_encode = pd.get_dummies(b2['Cities'], prefix='City_', dtype=int)
b2_encoded = pd.concat([b2, c_encode], axis=1)

b2_encoded
```

```python
# Ordinal Encoding
b3 = pd.DataFrame({'Education' : ['High School', 'Bachelor', 'Master', 'Bachelor', 'Master']})

order = {'High School': 0, 'Bachelor': 1, 'Master':2}
b3['edu_encoded'] = b3['Education'].map(order).astype(int)

b3
```

```python
# Vector
a = np.array([3, -1, 2])
b = np.array([4, 0, -2])
c = np.array([-6, 2, -4])

dot_ab = np.dot(a,b)
dot_ac = np.dot(a,c)

print(dot_ab)
print(dot_ac)

# Norms (Length or Magnitude of a Vector) 
l2_norm = np.linalg.norm(a)
print(l2_norm)

# Normalizing a Vector (Unit Vector)
unit_a = a/l2_norm
print(np.round(unit_a, 3))
```

```python
p1 = np.array([2, 3])
p2 = np.array([5, 7])
p3 = np.array([2, 10])

# Euclidean Distance (L2)
eu1 = np.linalg.norm(p1 - p2)
eu2 = np.linalg.norm(p2 - p3)
eu3 = np.linalg.norm(p1 - p3)

print("pair 1 =", np.round(eu1, 3))
print("pair 2 =", np.round(eu2, 3))
print("pair 3 =", np.round(eu3, 3))
print()

# Manhattan Distance (L1)
ma1 = np.linalg.norm(p1 - p2, ord=1)
ma2 = np.linalg.norm(p2 - p3, ord=1)
ma3 = np.linalg.norm(p1 - p3, ord=1)

print("pair 1 =", np.round(ma1, 3))
print("pair 2 =", np.round(ma2, 3))
print("pair 3 =", np.round(ma3, 3))
```
